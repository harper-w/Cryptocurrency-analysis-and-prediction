**4 Data Analysis**

In the data analysis, I will complete two tasks, the first is to analyse the price movements of the six major cryptocurrencies from 2018 to the present (converted to USD); the second is that I will analyse the relationship between the six major currencies. 

This is also a response to my Research Questions 1 and 2.

```python
# install the libraries

# pandas_datareader allows to create DataFrame by using data from the website
# especially real time financial data like stocks
import pandas_datareader as web

# using matplotlib.pyplot to plotting
import matplotlib.pyplot as plt

# using mplfiance to create financial plots
import mplfinance as mpf

# for data visualisation and analysis
import seaborn as sns

# datetime helps with date and time processing
import datetime

# data source: Yahoo Finance
import yfinance as yf

currency = "USD"  # set the currency comparator to US Dollar
metric = "Close"  # set the metric as the closing prices
```

```python
# set the biginning to 1st Janauary of 2018, as it the research question stated
start = datetime.datetime(2018, 1, 1)
# set the closing time to the current date
end = datetime.datetime.now()

# First Task
# ploting the the trend of six cryptocurrencies from Jan 2018 to now 

# pass the individual symbols(tickers) of the cryptocurrencies
crypto = ['BTC', 'ETH', 'LTC', 'XRP', 'DASH', 'SC']
# BTC: Bitcoin; ETH: Ethereum; LTC: Litecoin; XRP: Ripple; DASH: Dash; SC: Siacoin

# create an empty list for saving colnames later 
colnames = []
```

Since I want to put the prices of the six cryptocurrencies in a new dataframe, I will use a *for loop* and an *if statement*.

By the way, it is important to notice that the creation of *combined* below may occur overwrite or leading changes in the original data. In case of this may happen, I use a copy() function to replicate the data.

```python
# Create a dataframe containing the 2018 to date closing prices of six cryptocurrencies

cursor = True

for ticker in crypto: 
     # load data from Yahoo API; f refers to format Crpto Ticker to US Dollar Currency
    data = web.DataReader(f'{ticker}-{currency}', "yahoo", start, end)     
     #  cursor will be True at the first time in copying Bitrcoin data; then would be False the rest of five times 
    if cursor:
     # as mentioned in the markdown above, use copy() to avoid overwriting
     # name the replicate as combined      
        combined = data [[metric]].copy()  
     # append the Bitcoin ticker to the column name/label      
        colnames.append(ticker)      
        combined.columns = colnames
     # pause this loop, then let all the other cryptos go for the else part      
        cursor = False
    else:
     # continue adding the Close price of other cryptos 2018-now to the dataframe (first Bitcoin, then Ethereum, etc)    
        combined = combined.join(data[metric])
     # add another crypto names to the column names/labels   
        colnames.append(ticker)
        combined.columns = colnames
```

```python
# Now let's plot the dataframe above

# the value of the y-axis is expressed as a logarithmic function
# because the price disparity between different cryptocurrencies can be too large, we may not see any other currency than Bitcoin
plt.yscale('log')

# Plot the six cryptocurrencies into the chart in order
for ticker in crypto:
    plt.plot(combined[ticker], label = ticker)
    
plt.legend(loc="upper right")

plt.show()
```

```python
# Now let's plot a correlation heat map

# Because is 1. is coherent 2. is linearly correlated 3. satisfies a normal distribution
# so I choose Pearson correlation 
# pearson: pearson correlation
combined = combined.pct_change().corr(method = 'pearson')  

# draw a heatmap
sns.heatmap(combined, annot = True, cmap = 'coolwarm')

# change the figure size for all figures  
plt.rcParams['figure.figsize'] = (20, 20)

plt.show()
```

**5 Data Prediction**

In this section, I am going to predict the future day price of Bitcoin, which corresponds to Research Question 3. In this section, I will predict the future one-day price of Bitcoin, which corresponds to Research Question 3. I will train the model, as well as evaluate it, and then use it to make predictions.


```python
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.preprocessing import MinMaxScaler 
from tensorflow.keras.layers import Dense, Dropout, LSTM 
from tensorflow.keras.models import Sequential 

crypto_currency = 'BTC'
against_currency = 'USD'
```

```python
# the timefreme of the training data 
start = dt.datetime(2018, 1, 1)
end = dt.datetime.now()

data = web.DataReader(f'{crypto_currency}-{against_currency}', 'yahoo', start, end)
```

```python
# Prepare Data 
print(data.head()) 
```

```python
#  scale the range to between 0 and 1
scaler = MinMaxScaler(feature_range = (0,1)) 
# reshape the data form to fit1 the model
scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1,1))
```

```python
# Preparing the training data set 

# use the data of the past 60 days to predict the result of 61st day
prediction_days = 60

# x and y for supervised trainning
x_train, y_train = [], []
# in x_train list, it has n lists that contain the closing price within 60 days 
for x in range(prediction_days, len(scaled_data)):
    x_train.append(scaled_data[x - prediction_days:x, 0])
    y_train.append(scaled_data[x, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
# Create Neural Network
# Select the sequential model in Keras as it can deal with sequential integer
model = Sequential()
# using Sequetial to add the network structure layer by layer


# LSTM: Long Short Term Memory
# Attach three LSTM network into this model and set dropout to avoid overfitting
model.add(LSTM(units = 50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
# Final layer is dense layer which allow the model export only 1 value for each x data input
model.add(Dense(units=1))
# Compile the model by choosing the correct parameters
# adam is an optimization algorithm to interatively refresh the network weights on top of the training data
model.compile(optimizer='adam', loss='mean_squared_error')
# Training the model with the training dataset
model.fit(x_train, y_train, epochs=25, batch_size=32)
```

```python
# Testing and Evaluating the Model

# Create test dataset
# I want the model to predict the close price of BTC from 1/1/2020
test_start = dt.datetime(2020,1,1)
test_end = dt.datetime.now()

# The actual close price was extracted for validation
test_data = web.DataReader(f'{crypto_currency}-{against_currency}', 'yahoo', test_start, test_end)
actual_prices = test_data['Close'].values

# Create the input dataset that contain the close price 60 days before the test starting date for testing the model
total_dataset = pd.concat ((data['Close'], test_data['Close']), axis = 0)
model_inputs = total_dataset[len(total_dataset) - len(test_data) - prediction_days:].values
model_inputs = model_inputs.reshape(-1, 1)   # reshape this into column
model_inputs = scaler.fit_transform(model_inputs)
```

```python
# prepare the testing set for the model to predict the closing price
x_test = []

for x in range(prediction_days, len(model_inputs)):
    x_test.append(model_inputs[x-prediction_days:x, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1],1))
```

```python
# Make prediction 
# below is the closing price of the BTC predicted by model
prediction_prices = model.predict(x_test)

# The predicted closing price in USD by decoding from the model prediction
prediction_prices = scaler.inverse_transform(prediction_prices)
```

```python
# Plot the actual closing price and the closing price that the model predicts on a graph
plt.plot(actual_prices, color = 'black', label = 'Actual Prices')
plt.plot(prediction_prices, color = 'pink', label = 'Predicted Prices')
plt.title(f'{crypto_currency} price prediction')

plt.xlabel('Time')
plt.ylabel('Price')

plt.legend(loc = 'upper left')
plt.show()

# we can see here the per-trained model is reliable 
# it is not exactly accurate, but it is reliable
```

```python
# Actual Prediction for the next day

real_data = [model_inputs[len(model_inputs) + 1 - prediction_days: len(model_inputs) + 1, 0 ]]
real_data = np.array(real_data)
real_data = np.reshape(real_data, (real_data.shape[0], real_data.shape[1],1))

prediction = model.predict(real_data)
prediction = scaler.inverse_transform(prediction)
print(prediction)
```
